```python
import requests
from bs4 import BeautifulSoup

import pandas as pd

def imdb_original(title_type, start_date, end_date,user_rating):
        url = f'https://www.imdb.com/search/title/?title_type={title_type}&release_date={start_date},{end_date}&user_rating={user_rating},'
        response = requests.get(url)
        soup = BeautifulSoup(response.content)

        movies = soup.find_all("div", attrs = {"class":"lister-item-content"})
        movies_dict = {}
        movie_index = 0

        for movie in movies:
            film_header = movie.find_all("h3", attrs = {"class":"lister-item-header"})
            for item in film_header:
                movie_nr = item.find_all("span", attrs = {"class":"lister-item-index unbold text-primary"})[0].get_text().strip()
                title = item.find_all("a")[0].get_text().strip()
                year = item.find_all("span", attrs = {"class":"lister-item-year text-muted unbold"})[0].get_text().strip()
            rating = movie.find_all("div", attrs = {"class":"inline-block ratings-imdb-rating"})[0]["data-value"]
            runtime = movie.find_all("span", attrs = {"class":"runtime"})[0].get_text()
            genre = movie.find_all("span", attrs = {"class":"genre"})[0].get_text().strip()
            description = movie.find_all("p", attrs = {"class":"text-muted"})[1].get_text().strip()
            director_stars = movie.find_all("p")[2]
            director = director_stars.find_all("a")[0].get_text().strip()
            stars = ",".join([star.get_text().strip() for star in director_stars.find_all("a")[1:]])
            votes = movie.find_all("span", attrs = {"name":"nv"})[0]["data-value"]
            gross_earn = movie.find_all("span", attrs = {"name":"nv"})[-1]["data-value"]

            movies_dict[movie_index] = {"Movie NR":movie_nr,
                                    "Title":title,
                                    "Year": year,
                                    "Rating": rating,
                                    "Runtime": runtime,
                                    "Genre": genre,
                                    "Description": description,
                                    "Director": director,
                                    "Stars": stars,
                                    "Votes": votes,
                                    "Gross in $M": gross_earn}

            movie_index += 1
            df = pd.DataFrame.from_dict(movies_dict, orient="index").set_index("Movie NR")
        return df
imdb_original("feature","1990-01-01","1992-12-31","7.5")
```

```python
def get_all(title_type, start_date, end_date,user_rating):
    n_pages = int(input("how many pages you want?"))
    urls = []
    for i in range(1,n_pages):
        n_pages = 1 + 50*(i-1)
        url = f"https://www.imdb.com/search/title/?title_type={title_type}&release_date={start_date},{end_date}&user_rating={user_rating},&start={n_pages}&ref_=adv_nxt"
        urls.append(url)
        
    return urls
get_all("feature","1990-01-01","1992-12-31","7.5")
```
